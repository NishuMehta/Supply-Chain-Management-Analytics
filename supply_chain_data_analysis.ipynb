### DATA VISUALIZATION

#### 1. Products Sold for each Product Type

```python
prod_sold = data.groupby('Product type')['Number of products sold'].sum().reset_index().sort_values(by='Number of products sold',ascending=False)

plt.figure(figsize=(10,6))
plt.title('Number of Products Sold for each Product Type', fontsize=16)

plt.bar(prod_sold['Product type'],prod_sold['Number of products sold'], color='darkgreen')
plt.xlabel('Product Type')
plt.ylabel('Number of Products Sold')

plt.show()
```

#### 2. Revenue Generated Distribution by Product Type

```python
rev_prod = data.groupby('Product type')['Revenue generated'].sum()

plt.figure(figsize=(10,6))
plt.title('Distribution of Revenue Generated across Product Types', fontsize=16)

sns.boxplot(data=data, x='Product type', y='Revenue generated', color='green')
plt.show()
```

#### 3. Correlation between Price and Number of Products Sold

```python
plt.figure(figsize=(10,6))
plt.title('Correlation between Price and Products Sold', fontsize=16)

sns.scatterplot(
    x=data['Price'],
    y=data['Number of products sold'],
    alpha=1,
    color='green'
)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Number of Products Sold', fontsize=12)

plt.show()
```

#### 4. Customer Demographics

```python
cust_demo = data['Customer demographics'].value_counts()

plt.figure(figsize=(6,6))
plt.title('Customer Demographics Proportion',fontsize=16)

plt.pie(cust_demo, labels=cust_demo.index, autopct= '%1.1f%%', colors=['seagreen','lightgreen','green','darkgreen'])
plt.show()
```

#### 5. Number of Orders by Customer Demographics

```python
plt.figure(figsize = (10,6))
plt.suptitle('Customer Demographics Distribution', fontsize=16)

sns.countplot(data = data, x = 'Customer demographics', color='green', width = 0.4)
plt.show()
```

#### 6. Distribution of Lead Times

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Leading Times', fontsize=16)

sns.histplot(data['Lead times'], bins=12, kde=True, color='darkgreen')
plt.xlabel('Leading Times',fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

#### 7. Manufacturing Cost Trends

```python
data['Location'].value_counts()
```

```python
loc_mfg = data.groupby('Location')['Manufacturing costs'].mean()

plt.figure(figsize=(10,6))
plt.title('Manufacturing Costs', fontsize=16)

plt.plot(loc_mfg.index, loc_mfg.values, label='Kolkata', color ='green', marker = 'o')
plt.xlabel('Location',fontsize = 10)
plt.ylabel('Average Manufacturing Costs', fontsize = 10)

plt.show()
```

#### 8. Average Shipping Costs by Shipping Carrier

```python
ship_cost_carrier = data.groupby('Shipping carriers')['Shipping costs'].mean()

plt.figure(figsize=(10,6))
plt.title('Average Shipping Costs by Shipping Carrier', fontsize=16)

sns.barplot(x=ship_cost_carrier.index, y=ship_cost_carrier.values, color = 'green', width=0.5)
plt.xlabel('Shipping Carrier', fontsize=12)
plt.ylabel('Average Shipping Costs', fontsize=12)

plt.show()
```

#### 9 . Total Revenue by Location

```python
revenue_location = data.groupby('Location')['Revenue generated'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.title('Total Revenue Generated by Location', fontsize=16)

sns.barplot(x=revenue_location.index, y=revenue_location.values, color='green', width=0.4)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Total Revenue Generated', fontsize=12)

plt.show()
```

#### 9. Correlation between Defect Rates and Manufacturing Costs

```python
plt.figure(figsize=(10,6))
plt.title('Correlation between Defect Rate and Manufacturing Costs')

sns.scatterplot(
    x=data['Manufacturing costs'],
    y=data['Defect rates'],
    alpha=1,
    color='green'
)

plt.show()
```

#### 10. Count of Inspection Results

```python
inspections = data['Inspection results'].value_counts()

plt.figure(figsize=(10,6))
plt.title('Count of Inspection Results', fontsize=16)

sns.barplot(x=inspections.index, y=inspections.values, color='green', width=0.4)
plt.ylabel('Count', fontsize=12)
plt.show()
```

#### 11. Cost Distribution by Transportation Mode

```python
cust_data = data.groupby('Transportation modes')['Costs'].sum().reset_index()
colors = ['lightgreen', 'darkgreen', 'seagreen' ,'green']

plt.figure(figsize=(10,6))

plt.pie(x = cust_data['Costs'], autopct='%1.1f%%', colors=colors)
plt.title('Cost Distribution by Transportation Mode', fontsize=15)
plt.legend(labels = cust_data['Transportation modes'],loc="upper right", bbox_to_anchor=(1.1, 1))

plt.show()
```

#### 12. Defect Rates by Product Category

```python
defect_rate = data.groupby('Product type')['Defect rates'].mean().reset_index()
colors = ['pink','lightgreen','libhtblue','grey']

plt.figure(figsize=(10,6))
plt.bar(defect_rate['Product type'], defect_rate['Defect rates'], color='green', width = 0.5)
plt.title('Defect Rates by Product Category', fontsize=15, y = 1.05)
plt.xlabel('Product Type')
plt.ylabel('Defect Rates')

plt.show()
```

#### 13. Defect Rates by Transportation Mode

```python
defect_rate = data.groupby('Transportation modes')['Defect rates'].mean().reset_index()
colors = ['lightgreen', 'darkgreen', 'seagreen' ,'green']

plt.figure(figsize=(10,6))

plt.pie(x = defect_rate['Defect rates'], autopct='%1.1f%%', colors=colors)
plt.title('Defect Rates by Transportation Mode', fontsize=15)
plt.legend(labels = cust_data['Transportation modes'],loc="upper right", bbox_to_anchor=(1.1, 1))

plt.show()
```
